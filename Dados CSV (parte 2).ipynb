```
# Neste exercício, você deseja chamar a API last.fm para obter uma lista dos 
# melhores artistas da Espanha (método geo.getTopArtists).

# Sua função deve devolver o nome do cantor ou cantora número 1 na Espanha.

# Sua principal tarefa, aqui, é navegar pela estrutura de dados retornada pela 
# API last.fm para encontrar e retornar o melhor cantor da Espanha.

import json
import requests
import time
import pandas as pd

def get_top_artist(api_key):
  #Seu código vai aqui
  for page in range(1,2):
  #print(f'page:{page}')
    url = f"http://ws.audioscrobbler.com/2.0/?method=geo.gettopartists&api_key={api_key}&country=spain&format=json&limit=1000&page={page}"
    data = requests.get(url).text
    data = json.loads(data)
  for artist in data['topartists']['artist']:
    artist_name.append(artist['name'])
    artist_listeners.append(artist['listeners'])
  time.sleep(1)

  return data['topartists']['artist']['rank'==1]['name']

if __name__ == '__main__':
  artist_name =[]
  artist_listeners=[]
  api_key = 'c8eb2a5dfd4d0b7b9f8dbc250798d844'
  df = pd.DataFrame({'name':artist_name,'listeners':artist_listeners})
  df.sort_values(by=['listeners'],ascending=False).head()
  max_listeners = df['listeners'].max()
  df[df['listeners'] ==max_listeners]
  print(get_top_artist(api_key))
```

### 3.16 - Exercício de imputação (enviar pelo Canvas)

```
import pandas
import numpy

def imputation(filename):
    # Os dataframes do Pandas possuem um método chamado 'fillna(value)', que 
    # permite que seja passado um valor único para substituir qualquer NAs no
    # dataframe ou na série. Exemplo:
    #     dataframe['column'] = dataframe['column'].fillna(value)
    #
    # Usando a função numpy.mean, que calcula a média de um array, faça a 
    # imputação mais simples dos valores faltantes da coluna 'weight' do conjunto
    # de dados Lahman baseball.
    # Dataset: https://1drv.ms/u/s!Av8pSmXDlugUjc4-g-ZH_ENM6GkLVw

    baseball = pandas.read_csv(filename,encoding='ISO-8859-2')
  

    #Seu código vai aqui
    baseball['weight'] = baseball['weight'].fillna(185)
    baseball= numpy.mean(baseball['weight'])
    return baseball
```
