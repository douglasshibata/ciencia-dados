```
import requests
from bs4 import BeautifulSoup
import pandas as pd
import json

URL = 'https://pt.m.wikipedia.org/wiki/Lista_de_cursos_superiores_do_Brasil'
headers={
    'User-Agent': 'Mozilla/5.0 (Linux; Android 8.0; Pixel 2 Build/OPD3.170816.012) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/85.0.4183.83 Mobile Safari/537.36'
}
page = requests.get(URL,headers=headers)
soup = BeautifulSoup(page.content,'html.parser')
table = soup.findAll('ol')
table
cursos = list()
for i in table:
  li = i.find_all('li')
  row = [i.text.replace('\n','') for i in li]
  cursos.append(row)
cursos.pop(3)
cursos_1 = list()
for a in cursos:
  for b in a:
    cursos_1.append(b)
cursos_1.sort()
```

```
import requests
from bs4 import BeautifulSoup
import pandas as pd
import json

URL = 'https://www.ifb.edu.br/portal-do-aluno/196-estude-no-ifb/escolha-o-seu-curso/6007-cursos-tecnicos-integrado'
headers={
    'User-Agent': 'Mozilla/5.0 (Linux; Android 8.0; Pixel 2 Build/OPD3.170816.012) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/85.0.4183.83 Mobile Safari/537.36'
}
page = requests.get(URL,headers=headers)
soup = BeautifulSoup(page.content,'html.parser')
#table = soup.findAll('div',attrs={'informacao'})
table = soup.find_all('a')
table
tecs = list()
for i in table:
  tecs.append(i.getText())
del tecs[0:99]
del tecs[-7:]
tecnico = tecs + tec
del tecnico[-4:]
tecnico.sort()
supercursos = cursos_1 + tecnico
supercursos.sort()
supercursos
```

```
cursos_json = json.dumps(supercursos).encode('utf8')
with open('cursos.json', 'w', encoding='utf8') as json_file:
    json.dump(supercursos,json_file ,ensure_ascii=False)
```
