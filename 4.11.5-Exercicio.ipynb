```
import pandas as pd 
#!pip install -U pandasql
import pandasql
import matplotlib.pyplot as plt
import seaborn as sns
```

```
notas = pd.read_csv('ratings.csv')
notas.columns=['usuarioId','filmeId','nota','momento']
notas['nota'].unique()
notas
```

```
filmes = pd.read_csv("movies.csv")
filmes.columns = ['filmeId','titulo','genero']
filmes
```

```
notas.query('filmeId==1')
```

```
#result = pd.concat([notas, filmes], join='inner')
result = pd.merge(notas,filmes, on='filmeId')
result
```

```
notas_do_toy_story = result.query('filmeId==1')
print("Nota média do Toy Story %.2f" % notas_do_toy_story.nota.mean())
```

```
notas_do_pai_da_noiva_II = result.query('filmeId==5')
print("Nota média do Pai da Noiva II %.2f" % notas_do_pai_da_noiva_II.nota.mean())
```
